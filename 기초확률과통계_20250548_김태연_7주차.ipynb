```python
import matplotlib.pyplot as plt
from collections import Counter

dice_rolls = [(i, j) for i in range(1, 7) for j in range(1, 7)]
sums = [sum(roll) for roll in dice_rolls]

f_X = Counter(sums)
total_outcomes = len(dice_rolls)
f_X = {x: count / total_outcomes for x, count in f_X.items()}

plt.figure(figsize=(8, 6))
plt.plot(list(f_X.keys()), list(f_X.values()), 'bo')
plt.vlines(list(f_X.keys()), 0, list(f_X.values()), colors='blue')
plt.title('Probability Mass Function of X')
plt.xlabel('Sum of dice rolls (X)')
plt.ylabel('Probability')
plt.xticks(range(2, 13))
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
scores = [0, 1, 2, 3, 4]
probabilities_Y = [0.1, 0.2, 0.4, 0.2, 0.1]

expected_value_Y = sum(y * p for y, p in zip(scores, probabilities_Y))

variance_Y = sum((y - expected_value_Y) ** 2 * p for y, p in zip(scores, probabilities_Y))

print("기댓값 (E[Y]):", expected_value_Y)
print("분산 (Var[Y]):", variance_Y)
```
