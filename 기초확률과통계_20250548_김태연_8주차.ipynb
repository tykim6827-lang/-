```python
import matplotlib.pyplot as plt

joint_pmf_data = {
    (1, 1): 0.10,
    (1, 2): 0.20,
    (1, 3): 0.20,
    (2, 1): 0.30,
    (2, 2): 0.05,
    (2, 3): 0.15,
}

f_X = {}
f_Y = {}
for (x, y), p in joint_pmf_data.items():
    f_X[x] = f_X.get(x, 0.0) + p
    f_Y[y] = f_Y.get(y, 0.0) + p

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
xs = sorted(f_X.keys())
plt.plot(xs, [f_X[x] for x in xs], 'bo')
plt.vlines(xs, 0, [f_X[x] for x in xs], colors='blue')
plt.title('Marginal PMF of X')
plt.xlabel('X value')
plt.ylabel('Probability')
plt.xticks(xs)
plt.ylim(0, 1)

plt.subplot(1, 2, 2)
ys = sorted(f_Y.keys())
plt.plot(ys, [f_Y[y] for y in ys], 'ro')
plt.vlines(ys, 0, [f_Y[y] for y in ys], colors='red')
plt.title('Marginal PMF of Y')
plt.xlabel('Y value')
plt.ylabel('Probability')
plt.xticks(ys)
plt.ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

korean_scores = [75, 65, 50, 70, 80]
math_scores   = [65, 70, 70, 55, 60]

data = np.array([korean_scores, math_scores])
covariance = np.cov(data, bias=True)[0, 1]

print("Covariance:", covariance)
```
